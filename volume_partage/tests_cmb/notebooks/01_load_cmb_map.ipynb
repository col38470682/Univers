```python
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt

cmb_file = "/content/COM_CompMap_CMB-commrul_0256_R1.00.fits"

cmb_map = hp.read_map(cmb_file, field=0)

nside_out = hp.get_nside(cmb_map)

alm = hp.map2alm(cmb_map)
_, grad_theta, grad_phi = hp.alm2map_der1(alm, nside_out)

D_T = np.sqrt(grad_theta**2 + grad_phi**2)

print("NSIDE =", nside_out)
print("D_T min =", np.nanmin(D_T))
print("D_T max =", np.nanmax(D_T))
print("D_T mean =", np.nanmean(D_T))

hp.mollview(D_T, title="Gradient thermique CMB |∇T|")
plt.show()
```

```python
import numpy as np
import healpy as hp

theta = np.radians(90 - qso["dec"].values)
phi = np.radians(qso["ra"].values)

pix = hp.ang2pix(nside_out, theta, phi)

qso["D_T"] = D_T[pix]

print(qso.head())
print("Nombre de quasars :", len(qso))
```

```python
from scipy.stats import pearsonr, spearmanr
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

# Corrélations
pearson_r, pearson_p = pearsonr(qso["D_T"], qso["z"])
spearman_r, spearman_p = spearmanr(qso["D_T"], qso["z"])

print("========== CORRELATION ==========")
print("Pearson")
print("r =", pearson_r)
print("p =", pearson_p)

print()

print("Spearman")
print("r =", spearman_r)
print("p =", spearman_p)

# Ajustement linéaire
model = LinearRegression()

X = qso[["D_T"]]
y = qso["z"]

model.fit(X,y)

alpha = model.coef_[0]
beta = model.intercept_

print()
print("alpha =",alpha)
print("beta =",beta)

# graphique

plt.figure(figsize=(7,5))

plt.scatter(
    qso["D_T"],
    qso["z"],
    s=1,
    alpha=0.15
)

x=np.linspace(
    qso["D_T"].min(),
    qso["D_T"].max(),
    200
)

plt.plot(
    x,
    beta+alpha*x,
    linewidth=2
)

plt.xlabel("|∇T|")
plt.ylabel("Redshift z")
plt.title("SDSS Quasars : D_T vs z")

plt.show()
```

```python
from scipy.stats import spearmanr
import numpy as np
import pandas as pd

# ==========================
# Test de permutation
# ==========================

x = qso["D_T"].values
y = qso["z"].values

real_r, _ = spearmanr(x, y, nan_policy="omit")

Nperm = 2000
perm_r = np.zeros(Nperm)

rng = np.random.default_rng(42)

for i in range(Nperm):
    yp = rng.permutation(y)
    perm_r[i], _ = spearmanr(x, yp, nan_policy="omit")

p_perm = np.mean(np.abs(perm_r) >= abs(real_r))

print("="*50)
print("TEST DE PERMUTATION")
print("="*50)
print("Spearman réel :", real_r)
print("p permutation :", p_perm)
print("Moyenne permutations :", perm_r.mean())
print("Écart-type :", perm_r.std())

# ==========================
# Densité par quadrant
# ==========================

quad = []

for a, b in [(0,90),(90,180),(180,270),(270,360)]:

    sub = qso[(qso["ra"]>=a) & (qso["ra"]<b)]

    quad.append([
        f"{a}-{b}",
        len(sub),
        len(sub)/len(qso),
        sub["D_T"].mean(),
        sub["z"].mean()
    ])

res = pd.DataFrame(
    quad,
    columns=[
        "RA_quadrant",
        "N",
        "frac_N",
        "mean_D_T",
        "mean_z"
    ]
)

print("\n")
print("="*50)
print("RÉPARTITION PAR QUADRANT")
print("="*50)
print(res.to_string(index=False))
```

```python
from astropy.coordinates import SkyCoord
import astropy.units as u
import numpy as np

coords = SkyCoord(
    ra=qso["ra"].values*u.deg,
    dec=qso["dec"].values*u.deg,
    frame="icrs"
)

qso["gal_l"] = coords.galactic.l.deg
qso["gal_b"] = coords.galactic.b.deg

print(qso[["ra","dec","gal_l","gal_b"]].head())
```

```python
from scipy.stats import spearmanr

qso_masked = qso[np.abs(qso["gal_b"]) > 20].copy()

print("="*50)
print("MASQUE GALACTIQUE")
print("="*50)
print(f"Avant : {len(qso)}")
print(f"Après : {len(qso_masked)}")
print(f"Fraction conservée : {len(qso_masked)/len(qso):.2%}")

r, p = spearmanr(
    qso_masked["D_T"],
    qso_masked["z"]
)

print()
print("Spearman =", r)
print("p =", p)
```

```python
import numpy as np
from scipy.stats import spearmanr

x = qso_masked["D_T"].values
y = qso_masked["z"].values

real_r, _ = spearmanr(x, y)

Nperm = 2000
perm = np.zeros(Nperm)

rng = np.random.default_rng(42)

for i in range(Nperm):
    perm[i], _ = spearmanr(x, rng.permutation(y))

p_perm = np.mean(np.abs(perm) >= abs(real_r))

print("="*50)
print("PERMUTATION APRÈS MASQUE")
print("="*50)

print("Spearman réel :", real_r)
print("p permutation :", p_perm)
print("Moyenne :", perm.mean())
print("Écart-type :", perm.std())
```

```python
import numpy as np
import healpy as hp
from scipy.stats import spearmanr

# Blocs spatiaux larges
nside_block = 8

theta_b = np.radians(90 - qso_masked["dec"].values)
phi_b = np.radians(qso_masked["ra"].values)

block_id = hp.ang2pix(nside_block, theta_b, phi_b)

qso_masked["block"] = block_id

x = qso_masked["D_T"].values
y = qso_masked["z"].values
blocks = qso_masked["block"].values

real_r, _ = spearmanr(x, y)

Nperm = 1000
perm_r = np.zeros(Nperm)

rng = np.random.default_rng(0)

unique_blocks = np.unique(blocks)

for i in range(Nperm):
    y_shuffled = y.copy()

    for b in unique_blocks:
        idx = np.where(blocks == b)[0]
        if len(idx) > 1:
            y_shuffled[idx] = rng.permutation(y[idx])

    perm_r[i], _ = spearmanr(x, y_shuffled)

p_block = np.mean(np.abs(perm_r) >= abs(real_r))

print("Spearman réel :", real_r)
print("p permutation blocs spatiaux :", p_block)
print("Moyenne blocs :", perm_r.mean())
print("Sigma blocs :", perm_r.std())
print("Nombre de blocs utilisés :", len(unique_blocks))
```
